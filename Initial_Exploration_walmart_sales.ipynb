```python
#type_sales = df.groupby('Profit')['type_sales' = df.groupby('Profit')['Sales'].mean().reset_index()
sales_by_product_category = df.groupby('Product Category')['Sales'].mean().reset_index()
sales_by_product_category
sales_by_product_category.plot(kind='bar', x='Product Category', y='Sales', figsize=(10,6))
plt.xlabel('Product Category')
plt.ylabel('Sales')
plt.show()
```

```python


Region_wise_profit = df.groupby('Region')['Profit'].sum().reset_index()
Region_wise_profit
Region_wise_profit.plot(kind='bar', x='Region', y='Profit', figsize=(10,4),color='magenta')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.show()
```

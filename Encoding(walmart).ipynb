```python
import pandas as pd
import numpy as np

# Load the Excel file (replace 'your_file.xlsx' with your file name)
file_path = 'walmart Retail Data.xlsx'  # Update with your actual file path
data = pd.read_excel(file_path)

# Display the first few rows to understand the data
print("Dataset preview:")
print(data.head())

# Identify numerical columns
numerical_columns = data.select_dtypes(include=[np.number]).columns
print("\nNumerical columns:")
print(numerical_columns)

# Check for scaling needs by inspecting the range of numerical columns
scaling_recommendations = {}
for column in numerical_columns:
    col_min = data[column].min()
    col_max = data[column].max()
    col_mean = data[column].mean()
    col_std = data[column].std()

    if col_min >= 0 and col_max <= 1:
        scaling_recommendations[column] = "No scaling needed (already scaled)"
    elif col_mean < 10 and col_std < 10:
        scaling_recommendations[column] = "Standardization recommended"
    else:
        scaling_recommendations[column] = "MinMax Scaling recommended"

# Print recommendations for scaling
print("\nScaling recommendations:")
for column, recommendation in scaling_recommendations.items():
    print(f"{column}: {recommendation}")
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
import pandas as pd

# Load your dataset
file_path = 'walmart Retail Data.xlsx'  # Replace with your dataset path
data = pd.read_excel(file_path)

# Identify categorical columns
categorical_columns = data.select_dtypes(include=['object']).columns

# Count unique values in each categorical column
high_cardinality_features = {col: data[col].nunique() for col in categorical_columns}

# Filter columns with high cardinality (e.g., > 50 unique values)
threshold = 50
high_cardinality_features = {col: count for col, count in high_cardinality_features.items()
if count > threshold}

print("High Cardinality Features:")
print(high_cardinality_features)
```

```python
numeric_features = ['Customer Age', 'Order ID', 'Order Quantity', 'Profit', 'Row ID', 'Sales',
                    'Shipping Cost', 'Unit Price', 'Zip Code']
nominal_features = ['Customer Segment','Product Category','Product Container','Region','Ship Mode']
ordinal_features = ['Order Priority']
ordinal_Categories = ['High', 'Not Specified', 'Low', 'Critical', 'Medium']

high_card_features = ['City','Customer Name','Product Name']
```

```python
numeric_pipeline =Pipeline([('num_imputer',SimpleImputer(strategy='median')),('scaler',StandardScaler())])

numeric_pipeline.set_output(transform='pandas')
```

```python
ordinal_pipeline=Pipeline([('ordinal_imputer',SimpleImputer(strategy='most_frequent')),
 ('ordinal_encoder',OrdinalEncoder(categories=[ordinal_Categories]))])
ordinal_pipeline.set_output(transform='pandas')
```

```python
nominal_pipeline=Pipeline([('nominal_imputer',SimpleImputer(strategy='most_frequent')),
 ('nominal_encoder',OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
nominal_pipeline.set_output(transform='pandas')
```

```python
preprocessor=ColumnTransformer([('numeric_pipeline',numeric_pipeline,numeric_features),
                                ('ordinal_pipeline',ordinal_pipeline,ordinal_features),
                                ('nominal_pipeline',nominal_pipeline,nominal_features),
                                ('passthrough','passthrough',high_card_features)
                                ],remainder='drop')
preprocessor.set_output(transform='pandas')
```

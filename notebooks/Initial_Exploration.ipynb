Correlation b/w Sales Profit

```python
sale=['Item_MRP','Item_Visibility','Item_Outlet_Sales']
corr=df[sale].corr()
print(corr)
```

Item_MRP Item_Outlet_Sales are correlated showing positive correlation

```python
total_sales=df.groupby(['Item_Type','Outlet_Type', 'Outlet_Location_Type'])['Item_Outlet_Sales'].sum()  #performing group by where grouping 2 columns and performing sum on 3rd column
avg_sales=df.groupby(['Item_Type','Outlet_Type', 'Outlet_Location_Type'])['Item_Outlet_Sales'].mean()   #performing group by where grouping 2 columns and performing mean on 3rd column
print(total_sales) #calculating the total sales
print(avg_sales) #calculating average sales
```

```python
plots=df.groupby(['Item_Type','Outlet_Type', 'Outlet_Location_Type'])['Item_Outlet_Sales'].sum().reset_index()
print(plots) #plotting graph
```

```python
import matplotlib.pyplot as plt #importing library for visualizations
```

```python
type_sales = df.groupby('Item_Fat_Content')['Item_Outlet_Sales'].mean().reset_index()
print(type_sales)
type_sales.plot(kind='bar', x='Item_Fat_Content', y='Item_Outlet_Sales', figsize=(6,4))
plt.xlabel('Item_Fat_Content')
plt.ylabel('Item_Outlet_Sales')
plt.tight_layout()
plt.show()
```

Observations:
The Item_Outlet_Sales is highest for regular
The Item_Outlet_Sales is lowest for reg

```python
import matplotlib.pyplot as plt
```

```python
pie_data= df.groupby('Item_Type')['Item_Outlet_Sales'].sum().reset_index()
print('pie_data')
pie_data.plot(kind='bar', x='Item_Type', y='Item_Outlet_Sales', figsize=(10,4))
plt.xlabel('Item_Type')
plt.ylabel('Item_Outlet_Sales')
plt.tight_layout()
plt.show()
```

Observations:
Plot is between Item_Outlet_Sales and Item_Type
Fruits and Vegetable have the highest Item_Outlet_Sales.
Sea foods have the lowest Item_Outlet_Sales.

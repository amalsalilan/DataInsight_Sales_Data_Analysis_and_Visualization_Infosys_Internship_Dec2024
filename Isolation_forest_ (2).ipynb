```python
import pandas as pd
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
walmart = pd.read_excel(r'/content/drive/MyDrive/walmart Retail Data (1).xlsx')
walmart
```

```python


#df = pd.read_excel(r'C:\Users\Admin\Downloads\walmart Retail Data (1).xlsx')

features = ['Sales', 'Discount', 'Profit', 'Shipping Cost']
X = walmart[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = IsolationForest(contamination=0.01, random_state=42) # Adjust contamination as needed
model.fit(X_scaled)

walmart['anomaly'] = model.predict(X_scaled)
walmart['fraudulent'] = walmart['anomaly'].apply(lambda x: 1 if x == -1 else 0)

walmart['anomaly'] = [1 if x % 2 == 0 else 0 for x in range(len(walmart))] # Simulated data (replace if real labels exist)

precision = precision_score(walmart['anomaly'], walmart['fraudulent'])
recall = recall_score(walmart['anomaly'], walmart['fraudulent'])
f1 = f1_score(walmart['anomaly'], walmart['fraudulent'])

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")

# Step 8: Display flagged fraudulent transactions
fraudulent_transactions = walmart[walmart['fraudulent'] == 1]
print("\nFraudulent Transactions Detected:")
print(fraudulent_transactions[['Order ID', 'Sales', 'Discount', 'Profit', 'Shipping Cost', 'fraudulent']])
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from pingouin import partial_corr

# Select only numerical columns
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# Initialize an empty DataFrame to store partial correlations
columns = numerical_df.columns
partial_corr_matrix = pd.DataFrame(index=columns, columns=columns, dtype=float)

# Compute partial correlations for each pair of variables
for i in range(len(columns)):
    for j in range(len(columns)):
        if i != j:  # Avoid self-correlation
            # Control for all other variables except the current pair
            controlled_vars = list(columns.difference([columns[i], columns[j]]))
            result = partial_corr(data=numerical_df, x=columns[i], y=columns[j], covar=controlled_vars)
            partial_corr_matrix.loc[columns[i], columns[j]] = result['r'].values[0]
        else:
            partial_corr_matrix.loc[columns[i], columns[j]] = 1  # Self-correlation is 1

# Plot the heatmap for partial correlations
plt.figure(figsize=(10, 8))
sns.heatmap(partial_corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Partial Correlation Matrix")
plt.show()
```

```python
import numpy as np
from statsmodels.tsa.stattools import adfuller

# Function to check stationarity
def check_stationarity(series, column_name):
    series = series.dropna()  # Remove NaN values
    result = adfuller(series)

    print(f"Results of ADF Test for {column_name}:")
    print(f"ADF Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    print(f"Critical Values: {result[4]}")

    if result[1] < 0.05:
        print(f"The {column_name} series is stationary (reject H0).")
        return True
    else:
        print(f"The {column_name} series is non-stationary (fail to reject H0).")
        return False

# Function to make the series stationary
def make_stationary(series, column_name):
    print(f"Processing column: {column_name}")

    # First differencing
    series_diff = series.diff().dropna()
    if check_stationarity(series_diff, f"{column_name}_diff_1"):
        return series_diff

 # Second differencing if the first differencing isn't sufficient
    series_diff_2 = series_diff.diff().dropna()
    if check_stationarity(series_diff_2, f"{column_name}_diff_2"):
        return series_diff_2

    # Log transformation (useful for series with large values)
    series_log = np.log(series + 1).replace(-np.inf, np.nan).dropna()
    series_log_diff = series_log.diff().dropna()
    if check_stationarity(series_log_diff, f"{column_name}_log_diff"):
        return series_log_diff

    # If all fail
    print(f"Could not achieve stationarity for {column_name}.")
    return series

# Example usage with 'Customer Age'
df['Customer_Age_stationary'] = make_stationary(df['Customer Age'], 'Customer Age')

# Recheck stationarity of the transformed series
check_stationarity(df['Customer_Age_stationary'], 'Customer_Age_stationary')

```

```python
# performing ADF test for whole dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

# Function to perform ADF test and plot the results
def check_stationarity_with_plot(series, column_name, step_name):
    series = series.dropna()  # Remove NaN values for ADF test

    # Perform ADF test
    result = adfuller(series)
    adf_stat = result[0]
    p_value = result[1]
    critical_values = result[4]

    # Plot the series
    plt.figure(figsize=(12, 6))
    plt.plot(series, label=f"{column_name} ({step_name})", color='blue')
    plt.axhline(y=0, color='red', linestyle='--', linewidth=1)
    plt.title(f"ADF Test: {column_name} ({step_name})", fontsize=16)
    plt.xlabel("Index", fontsize=14)
    plt.ylabel("Values", fontsize=14)
    plt.legend(fontsize=12)
    plt.grid()

    # Display ADF test results as text on the plot
    textstr = f"ADF Statistic: {adf_stat:.4f}\nP-Value: {p_value:.4f}\n"
    for key, value in critical_values.items():
        textstr += f"Critical Value ({key}): {value:.4f}\n"
    plt.gca().text(0.02, 0.85, textstr, transform=plt.gca().transAxes,
                   fontsize=12, bbox=dict(facecolor='white', alpha=0.7))
    plt.show()

    # Determine if the series is stationary
    if p_value < 0.05:
        print(f"The {column_name} series is stationary (reject H0).")
        return True
    else:
        print(f"The {column_name} series is non-stationary (fail to reject H0).")
        return False

# Function to make a series stationary with visuals
def make_stationary_with_plots(series, column_name):
    print(f"Analyzing column: {column_name}")

    # Step 1: Original Series
    is_stationary = check_stationarity_with_plot(series, column_name, "Original")
    if is_stationary:
        return series

    # Step 2: First Differencing
    series_diff = series.diff()
    is_stationary = check_stationarity_with_plot(series_diff, column_name, "First Differencing")
    if is_stationary:
        return series_diff

    # Step 3: Log Transformation + Differencing
    series_log = np.log(series + 1)
    series_log_diff = series_log.diff()
    is_stationary = check_stationarity_with_plot(series_log_diff, column_name, "Log Transformation + Differencing")
    if is_stationary:
        return series_log_diff

    print(f"Could not achieve stationarity for {column_name}.")
    return series

# Example usage with 'Customer Age'
df['Customer_Age_stationary'] = make_stationary_with_plots(df['Customer Age'], 'Customer Age')
```

```python
# Detecting Outliers for whole dataset
import pandas as pd

# Load your dataset
file_path = "/content/walmart Retail Data.xlsx"  # Adjust to your file path
df = pd.read_excel(file_path)

# Detect numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Dictionary to store columns with outliers
outliers_info = {}

# IQR method for outlier detection
for column in numerical_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Detect outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    if not outliers.empty:
        outliers_info[column] = outliers

# Print columns with outliers
if outliers_info:
    print("Columns with outliers and their counts:")
    for column, outlier_data in outliers_info.items():
        print(f"{column}: {len(outlier_data)} outliers")
else:
    print("No outliers detected in the dataset.")

# If you want to inspect the outliers for a specific column, uncomment the next line:
# print(outliers_info['Your_Column_Name'])
#outliers detection of walmart dataset
```

```python
# performing t-SNE (t-distributed stochastic Neighbor Embedding) in 3D
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import numpy as np

# Load the data
data = pd.read_excel('/content/walmart Retail Data.xlsx')

# Define features and target
features = data.drop(['Customer Age','Discount','Profit','Sales','Shipping Cost','Unit Price'],axis=1)
targets = 'Sales'

# Copy the data for normalization
data_norm = data.copy()

# Initialize StandardScaler
sc = StandardScaler()

# Identify numerical columns in the features DataFrame
numerical_features = features.select_dtypes(include=['float64', 'int64']).columns

# Fit and transform the numerical columns, then assign them back to the copy
data_norm[numerical_features] = sc.fit_transform(data[numerical_features])

# Initialize SimpleImputer to handle NaNs
imputer = SimpleImputer(strategy='mean')

# Extract numerical column names from the features DataFrame
numerical_feature_columns = features.select_dtypes(include=['float64', 'int64']).columns

# Fit and transform the data using only the numerical columns using the imputer first.
data_norm[numerical_feature_columns] = imputer.fit_transform(data_norm[numerical_feature_columns])

# Initialize t-SNE with 3 components for 3D
tsne = TSNE(learning_rate=500, n_components=3, random_state=42)

# Fit and transform the data using only the numerical columns
x_tsne = tsne.fit_transform(data_norm[numerical_feature_columns])

# Print the shape to verify that it was successful
print(x_tsne.shape)

y_tsne=data[targets]

pd.DataFrame(x_tsne)

# 3D plot
fig = plt.figure(figsize=(16, 11))
ax = fig.add_subplot(111, projection='3d')  # Create a 3D plot

# Scatter plot in 3D
scatter = ax.scatter(x_tsne[:, 0], x_tsne[:, 1], x_tsne[:, 2], c=y_tsne, cmap='flag_r')

# Add colorbar for the scatter plot
colorbar = plt.colorbar(scatter, ax=ax, label='Sales')
colorbar.set_label('Sales', color='red')

# Set labels and title
ax.set_xlabel('Dimension 1')
ax.set_ylabel('Dimension 2')
ax.set_zlabel('Dimension 3')
ax.set_title('3D t-SNE Visualization of Data')

plt.show()
```

```python
# Detecting outliers using z-score
# Data points with a z-score greater than 3 or less than -3 may be outliers
import pandas as pd
import numpy as np
import plotly.express as px

# Load the dataset
df = pd.read_excel("walmart Retail Data.xlsx")

# Extract "Order Date" and "Sales" columns
df = df[['Order Date', 'Sales']]

# Calculate Z-scores for the "Sales" column
df['Z-Score'] = (df['Sales'] - df['Sales'].mean()) / df['Sales'].std()

# Identify outliers where the absolute Z-score is greater than 3
outliers = df[np.abs(df['Z-Score']) > 3]

# Plot the sales data and highlight outliers
fig = px.scatter(df, x='Order Date', y='Sales', title="Sales with Outliers Highlighted")
fig.add_scatter(x=outliers['Order Date'], y=outliers['Sales'], mode='markers',
                marker=dict(color='red', size=10), name='Outliers')
fig.show()
```

```python
# outliers by IQR method
import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_excel("walmart Retail Data.xlsx")

# Extract "Order Date" and "Sales" columns
df = df[['Order Date', 'Sales']]

# Calculate IQR
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Sales'] < lower_bound) | (df['Sales'] > upper_bound)]

# Plot sales and highlight outliers
fig = px.scatter(df, x='Order Date', y='Sales', title="Sales with Outliers Highlighted")
fig.add_scatter(x=outliers['Order Date'], y=outliers['Sales'], mode='markers',
                marker=dict(color='red', size=10), name='Outliers')
fig.show()
```

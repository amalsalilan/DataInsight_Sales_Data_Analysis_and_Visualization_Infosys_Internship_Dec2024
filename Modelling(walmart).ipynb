```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the training and testing datasets
train_data = pd.read_excel('/mnt/data/Training_Data.xlsx')
test_data = pd.read_excel('/mnt/data/Testing_Data.xlsx')

# Convert 'Order Date' to datetime format
train_data['Order Date'] = pd.to_datetime(train_data['Order Date'])
test_data['Order Date'] = pd.to_datetime(test_data['Order Date'])

# Sort data by 'Order Date'
train_data = train_data.sort_values('Order Date')
test_data = test_data.sort_values('Order Date')

# Convert 'Order Date' to ordinal for modeling
train_data['Order Date Ordinal'] = train_data['Order Date'].map(pd.Timestamp.toordinal)
test_data['Order Date Ordinal'] = test_data['Order Date'].map(pd.Timestamp.toordinal)

# Features and target for training
X_train = train_data[['Order Date Ordinal']]
y_train = train_data['Sales']

# Train a simple Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on the test set
X_test = test_data[['Order Date Ordinal']]
y_test = test_data['Sales']
y_pred = model.predict(X_test)

# Convert the ordinal back to datetime for plotting
test_data['Predicted Sales'] = y_pred
test_data = test_data.sort_values('Order Date')

# Visualize the predictions vs actual values
plt.figure(figsize=(12, 6))
plt.plot(test_data['Order Date'], test_data['Sales'], label='Actual Sales', marker='o', color='blue')
plt.plot(test_data['Order Date'], test_data['Predicted Sales'], label='Predicted Sales', marker='x', color='red')
plt.title('Actual vs Predicted Sales')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
import plotly.graph_objects as go

# Load the training dataset
train_data = pd.read_excel('/mnt/data/Training_Data.xlsx')

# Prepare the data: Select 'Order Date' and 'Sales' columns, convert 'Order Date' to datetime, and set it as the index
train_data = train_data[['Order Date', 'Sales']]
train_data['Order Date'] = pd.to_datetime(train_data['Order Date'])
train_data.set_index('Order Date', inplace=True)

# Resample the data to monthly frequency (or adjust as per your data)
monthly_sales = train_data['Sales'].resample('ME').mean()

# Fit the SARIMA model (adjust the order and seasonal_order as needed)
model = SARIMAX(monthly_sales,
                order=(1, 1, 1),          # (p, d, q)
                seasonal_order=(1, 1, 1, 12))  # (P, D, Q, s), where s=12 for monthly data
sarima_fit = model.fit()

# Forecast the next 12 months
forecast = sarima_fit.get_forecast(steps=12)
forecast_index = pd.date_range(monthly_sales.index[-1] + pd.DateOffset(1), periods=12, freq='ME')
forecast_values = forecast.predicted_mean
conf_int = forecast.conf_int()

# Plot using Plotly
fig = go.Figure()

# Historical Sales
fig.add_trace(go.Scatter(
    x=monthly_sales.index,
    y=monthly_sales,
    mode='lines',
    name='Historical Sales',
    line=dict(color='blue')
))

# Forecasted Sales
fig.add_trace(go.Scatter(
    x=forecast_index,
    y=forecast_values,
    mode='lines',
    name='Forecasted Sales',
    line=dict(color='orange')
))

# Confidence Interval
# Convert the forecast_index to a pandas Series for concatenation
forecast_index_series = pd.Series(forecast_index)
fig.add_trace(go.Scatter(
    x=pd.concat([forecast_index_series, forecast_index_series[::-1]]),
    y=pd.concat([conf_int.iloc[:, 0], conf_int.iloc[:, 1][::-1]]),
    fill='toself',
    fillcolor='rgba(255, 165, 0, 0.3)',
    line=dict(color='rgba(255, 165, 0, 0)'),
    name='Confidence Interval'
))

# Layout customization
fig.update_layout(
    title="Sales Forecast with SARIMA Model",
    xaxis_title="Date",
    yaxis_title="Sales",
    legend=dict(x=0.01, y=0.99),
    template="plotly_white",
    width=900,
    height=500
)

fig.show()

# Print forecasted values
print("Forecasted Sales Values:")
print(forecast_values)

```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import pandas as pd
import numpy as np

# Load the dataset
file_path = '/mnt/data/Training_Data.xlsx'  # Update with your file path
df = pd.read_excel(file_path)
df = df[['Order Date', 'Sales']]

# Preprocess the data
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.set_index('Order Date', inplace=True)
monthly_sales = df['Sales'].resample('ME').mean()

# Handle missing values
monthly_sales = monthly_sales.dropna()

# Scale the data
scaler = MinMaxScaler()
monthly_sales_scaled = scaler.fit_transform(monthly_sales.values.reshape(-1, 1))

# Train the SARIMA model on scaled data
model = SARIMAX(monthly_sales_scaled,
                order=(1, 1, 1),
                seasonal_order=(1, 1, 1, 12))
sarima_fit = model.fit()

# Forecasting (scaled values)
forecast_scaled = sarima_fit.get_forecast(steps=12)
forecast_scaled_values = forecast_scaled.predicted_mean
forecast_index = pd.date_range(monthly_sales.index[-1], periods=12, freq='ME')

# Inverse transform the scaled forecast back to original scale
forecast_values = scaler.inverse_transform(forecast_scaled_values.reshape(-1, 1))

# Calculate confidence intervals (scaled and inverse transform)
conf_int_scaled = forecast_scaled.conf_int()
conf_int = scaler.inverse_transform(conf_int_scaled)

# Calculate errors
actual_values = monthly_sales[-12:].values  # Assuming last 12 months are used for testing
forecast_original_values = forecast_values.flatten()

# Mean Squared Error (MSE)
mse = mean_squared_error(actual_values, forecast_original_values)

# Mean Absolute Error (MAE)
mae = mean_absolute_error(actual_values, forecast_original_values)

# Print the results
print("Forecasted Values (Original Scale):")
print(forecast_original_values)
print("\nMean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Load the dataset
data = pd.read_excel('/mnt/data/Training_Data.xlsx')

# Convert "Order Date" to datetime and aggregate sales by month
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Assuming 'Sales' is the original sales column
if 'Log Sales' not in data.columns:
    data['Log Sales'] = np.log(data['Sales'])

monthly_sales = data.groupby(data['Order Date'].dt.to_period('M')).sum(numeric_only=True)['Log Sales']
monthly_sales = monthly_sales.to_timestamp()

# Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, label='Monthly Sales', color='blue')
plt.title('Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()

# Split data into training and testing sets
train = monthly_sales[:-12]  # Use all but the last 12 months for training
test = monthly_sales[-12:]  # Use the last 12 months for testing

# Fit the SARIMAX model
sarima_model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
result = sarima_model.fit()

# Step 1: Forecast using the model (get_forecast)
forecast_steps = 12  # Forecast the next 12 months
forecast_result = result.get_forecast(steps=forecast_steps)

# Step 2: Get forecasted values (predicted_mean) and confidence intervals
forecast_values = forecast_result.predicted_mean
conf_int = forecast_result.conf_int()

# Step 3: Create a date range for the forecast period
forecast_index = pd.date_range(start=train.index[-1], periods=forecast_steps + 1, freq='M')[1:]

# Step 4: Plotting the forecast with confidence intervals
plt.figure(figsize=(12, 6))
plt.plot(train, label='Training Data', color='blue')
plt.plot(test, label='Test Data', color='green')
plt.plot(forecast_index, forecast_values, label='Forecast', color='red')

# Plot the confidence intervals
plt.fill_between(forecast_index,
                 conf_int.iloc[:, 0],  # Lower bound of confidence interval
                 conf_int.iloc[:, 1],  # Upper bound of confidence interval
                 color='pink', alpha=0.3)

plt.title('Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()

# Step 5: Evaluation metrics
# Align test and forecast_values for comparison
mae = mean_absolute_error(test, forecast_values)
mse = mean_squared_error(test, forecast_values)
rmse = np.sqrt(mse)

# Print evaluation metrics
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
# Holt-Winters Exponential Smoothing

import pandas as pd
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from prophet import Prophet
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel('/mnt/data/Training_Data.xlsx')

# Prepare the data: Select 'Order Date' and 'Sales' columns
df = df[['Order Date', 'Sales']]
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.rename(columns={'Order Date': 'ds', 'Sales': 'y'}, inplace=True)  # Rename columns for Prophet compatibility

# Use Prophet to clean the data (detect trends and remove anomalies)
prophet = Prophet()
prophet.fit(df)

# Generate cleaned data using Prophet's trend prediction
future_dates = prophet.make_future_dataframe(periods=0)  # No additional future dates, just clean existing data
cleaned_data = prophet.predict(future_dates)
df['y'] = cleaned_data['trend']  # Use the trend column as the cleaned sales data

# Resample cleaned data to monthly frequency (12 months seasonality)
df.set_index('ds', inplace=True)
monthly_sales = df['y'].resample('M').sum()

# Ensure monthly_sales is positive (add a small positive value if necessary)
monthly_sales[monthly_sales <= 0] = 1e-6

# Fit Holt-Winters Exponential Smoothing model with monthly seasonality (12 periods)
hw_model = ExponentialSmoothing(monthly_sales,
                                 seasonal='mul',
                                 seasonal_periods=12).fit()

# Forecast sales up to the end of 2017 (365 days is removed, predict until December 31, 2017)
forecast_steps = pd.date_range(start=monthly_sales.index[-1] + pd.DateOffset(1),
                               end="2016-12-31", freq='ME')
forecast = hw_model.forecast(len(forecast_steps))

# Plot the historical data, cleaned trend, and forecast
plt.figure(figsize=(14, 7))
plt.plot(monthly_sales, label='Cleaned Historical Sales', color='blue')
plt.plot(forecast_steps, forecast, label='Forecasted Sales (2017)', color='green', linestyle='--')
plt.title("Holt-Winters Forecasting (Cleaned with Prophet) - Sales Forecast until 2017")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()

```

```python
from prophet import Prophet
import pandas as pd

# Load your data (replace 'your_file.xlsx' with your actual file path)
data = pd.read_excel("walmart Retail Data.xlsx")

# Preprocess the data for Prophet
time_series_data = data[['Order Date', 'Sales']].copy()
time_series_data.rename(columns={'Order Date': 'ds', 'Sales': 'y'}, inplace=True)

# Ensure the 'ds' column is in datetime format and sort the data by date
time_series_data['ds'] = pd.to_datetime(time_series_data['ds'])
time_series_data = time_series_data.sort_values('ds')

# Aggregate sales by date (if multiple entries exist for the same date)
time_series_data = time_series_data.groupby('ds').sum().reset_index()

# Initialize and fit the Prophet model
model = Prophet()
model.fit(time_series_data)

# Create a DataFrame for future dates
future = model.make_future_dataframe(periods=365)  # Predict for the next 365 days

# Forecast sales
forecast = model.predict(future)

# Plot the forecast
fig = model.plot(forecast)
```

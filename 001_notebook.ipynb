```python
from sklearn.impute import SimpleImputer
```

```python
numerical_features = ["Estimated_Days", "Actual_Days"]
categorical_features = ["Shipping_Method", "Weather_Condition", "Region", "Courier", "Priority_Level"]
```

```python
num_imputer=SimpleImputer(strategy='median')
df[numerical_features]=num_imputer.fit_transform(df[numerical_features])

cat_imputer=SimpleImputer(strategy='most_frequent')
df[categorical_features]=cat_imputer.fit_transform(df[categorical_features])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')
```

```python
df
```

```python
df['Shipping_Method']
```

```python
ohe_transformed_df=ohe.fit_transform(df[['Shipping_Method','Weather_Condition','Region','Courier']])
```

```python
ohe_transformed_df
```

```python
df=pd.concat([df,ohe_transformed_df],axis=1).drop(columns=['Shipping_Method','Weather_Condition','Region','Courier'])
```

Relationships between the attributes

```python
# Compute the correlation matrix for all numerical columns
correlation_matrix = df.select_dtypes(include= ['number']).corr()
correlation_matrix
```

```python
correlation_matrix = df[['Item_MRP', 'Item_Visibility', 'Item_Outlet_Sales']].corr()
correlation_matrix
```

Item_MRP and Item_Outlet_Sales have a moderate positive correlation (0.567574), meaning higher prices might be associated with higher sales.

```python
sales = df['Item_Outlet_Sales'].describe()
sales
```

The goal is to examine items with zero visibility and analyze their identifiers, types, and sales to understand their behavior despite no visibility.

```python
zero_visibility_items = df[df['Item_Visibility'] == 0]
print(zero_visibility_items[['Item_Identifier', 'Item_Type', 'Item_Outlet_Sales']])
zero_visibility_items.shape
```

```python
visibility_comparison = df.groupby(df['Item_Visibility'] == 0)['Item_Outlet_Sales'].count()
visibility_comparison
```

The correlation is -0.128625, indicating a weak negative relationship. This suggests that as visibility increases, sales tend to decrease slightly, but the effect is minimal.

Missing Data Exploration

```python
# missing_data = df[['Item_Weight', 'Outlet_Size']].isnull().sum()
# missing_data
missing_values = df.isnull().sum()
missing_values
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_percentage
```

```python
missing_item_weight = df[df['Item_Weight'].isnull()]
print(missing_item_weight[['Item_Identifier', 'Item_Type', 'Outlet_Type']])

missing_item_weight_by_type = missing_item_weight.groupby('Item_Type').size()
print(missing_item_weight_by_type)

```

```python
# Fill missing Item_Weight with the median value of the Item_Type group
df['Item_Weight'] = df.groupby('Item_Type')['Item_Weight'].transform(lambda x: x.fillna(x.median()))


# Fill missing Outlet_Size with the most frequent value in Outlet_Type group
df['Outlet_Size'] = df.groupby('Outlet_Type')['Outlet_Size'].transform(lambda x: x.fillna(x.mode()[0]))

df.isnull().sum()
```

```python
# Visualization 2: Correlation heatmap of numerical features
plt.figure(figsize=(12, 8))
correlation_matrix = df[['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Item_Outlet_Sales']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Visualization 3: Item Visibility vs. Sales
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Item_Visibility', y='Item_Outlet_Sales', hue='Item_Fat_Content', alpha=0.6)
plt.title('Item Visibility vs. Item Outlet Sales', fontsize=16)
plt.xlabel('Item Visibility', fontsize=12)
plt.ylabel('Item Outlet Sales', fontsize=12)
plt.legend(title='Fat Content')
plt.tight_layout()
plt.show()
```

```python

# Visualization 4: Sales over Establishment Year
plt.figure(figsize=(10, 6))
df.groupby('Outlet_Establishment_Year')['Item_Outlet_Sales'].sum().plot(kind='bar', color='skyblue')
plt.title('Total Sales by Outlet Establishment Year', fontsize=16)
plt.xlabel('Establishment Year', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Visualization 5: Distribution of Item Fat Content
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Item_Fat_Content', palette='muted', order=df['Item_Fat_Content'].value_counts().index)
plt.title('Distribution of Item Fat Content', fontsize=16)
plt.xlabel('Item Fat Content', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.tight_layout()
plt.show()
```

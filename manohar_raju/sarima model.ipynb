```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a DataFrame
data = pd.read_excel('C:/Users/manu/OneDrive/Documents/spin/walmart Retail Data.xlsx')

# Assuming 'Order Date' and 'Sales' exist in the original columns
df = data[["Order Date", "Sales"]]

# Rename the selected columns
df.columns = ["Order Date", "Sales"]

# Convert 'Order Date' to datetime and extract year and month
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')  # Handle invalid dates
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

# Define a function to get the season based on the month
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
    else:
        return 'Unknown'  # Handle unexpected month values

# Apply the function to create a 'Season' column
df['Season'] = df['Month'].apply(get_season)

# Group by Year and Season, then sum the sales
seasonal_sales = df.groupby(['Year', 'Season'])['Sales'].sum().unstack()

# Plot the data
plt.figure(figsize=(10, 6))
for season in seasonal_sales.columns:
    plt.plot(seasonal_sales.index, seasonal_sales[season], label=season, marker='o')

plt.title('Year-over-Year Sales by Season')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.legend(title='Season')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Step 1: Load the Excel file
file_path = 'C:/Users/manu/OneDrive/Documents/spin/walmart Retail Data.xlsx'
data = pd.read_excel(file_path)

# Step 2: Check the initial structure of the dataset
print("Initial Data Shape:", data.shape)
print("\nColumns:\n", data.columns)
print("\nData Types:\n", data.dtypes)
print("\nFirst Few Rows:\n", data.head())

# Step 3: Select and Rename Columns
# Ensure the columns exist in the dataset
if "Order Date" in data.columns and "Sales" in data.columns:
    data = data[["Order Date", "Sales"]]  # Select relevant columns
    data.columns = ["Order Date", "Sales"]  # Rename columns
else:
    raise KeyError("Columns 'Order Date' and 'Sales' not found in the dataset!")

# Step 4: Convert Data Types
# Convert 'Order Date' to datetime and 'Sales' to numeric
data["Order Date"] = pd.to_datetime(data["Order Date"], errors="coerce")
data["Sales"] = pd.to_numeric(data["Sales"], errors="coerce")

# Step 5: Handle Missing Values
# Drop rows with missing values in critical columns
data.dropna(subset=["Order Date", "Sales"], inplace=True)

# Step 6: Remove Duplicates
data.drop_duplicates(inplace=True)

# Step 7: Handle Outliers
# Example: Remove extreme sales values (beyond the 1st and 99th percentiles)
sales_min = data["Sales"].quantile(0.01)
sales_max = data["Sales"].quantile(0.99)
data = data[(data["Sales"] >= sales_min) & (data["Sales"] <= sales_max)]

# Step 8: Feature Engineering
# Add Year and Month columns based on 'Order Date'
data["Year"] = data["Order Date"].dt.year
data["Month"] = data["Order Date"].dt.month

# Step 9: Normalize Sales
scaler = MinMaxScaler()
data["Sales_Normalized"] = scaler.fit_transform(data[["Sales"]])

# Step 10: Verify the Cleaned Dataset
print("\nCleaned Data Shape:", data.shape)
print("\nPreview of Processed Data:\n", data.head())

# Step 11: Save the Processed Data
processed_file_path = 'C:/Users/manu/OneDrive/Documents/spin/walmart_processed_cleaned.xlsx'
data.to_excel(processed_file_path, index=False)
print(f"\nProcessed data saved to: {processed_file_path}")
```

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# Step 1: Load the dataset
file_path = 'C:/Users/manu/OneDrive/Documents/spin/walmart_processed_cleaned.xlsx'
data = pd.read_excel(file_path)

# Step 2: Select and Rename Columns (if needed)
if "Order Date" in data.columns and "Sales" in data.columns:
    data = data[["Order Date", "Sales"]]  # Select relevant columns
    data.columns = ["Order Date", "Sales"]  # Rename columns
else:
    raise KeyError("Columns 'Order Date' and 'Sales' not found in the dataset!")

# Step 3: Convert 'Order Date' to datetime and sort data
data["Order Date"] = pd.to_datetime(data["Order Date"])
data.sort_values("Order Date", inplace=True)

# Step 4: Set 'Order Date' as the index
data.set_index("Order Date", inplace=True)

# Step 5: Extract the 'Sales' column for analysis
sales_data = data["Sales"]

# Step 6: Perform the Augmented Dickey-Fuller test
adf_result = adfuller(sales_data)

# Step 7: Display the results
print("ADF Statistic:", adf_result[0])
print("p-value:", adf_result[1])
print("Critical Values:")
for key, value in adf_result[4].items():
    print(f"   {key}: {value}")

# Step 8: Interpret the results
if adf_result[1] <= 0.05:
    print("\nThe data is stationary (reject the null hypothesis).")
else:
    print("\nThe data is not stationary (fail to reject the null hypothesis).")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

# Step 1: Load the dataset
file_path = 'C:/Users/manu/OneDrive/Documents/spin/walmart_processed_cleaned.xlsx'
data = pd.read_excel(file_path)

# Step 2: Select and Rename Columns (if needed)
if "Order Date" in data.columns and "Sales" in data.columns:
    data = data[["Order Date", "Sales"]]  # Select relevant columns
    data.columns = ["Order Date", "Sales"]  # Rename columns
else:
    raise KeyError("Columns 'Order Date' and 'Sales' not found in the dataset!")

# Step 3: Convert 'Order Date' to datetime and sort data
data["Order Date"] = pd.to_datetime(data["Order Date"])
data.sort_values("Order Date", inplace=True)

# Step 4: Set 'Order Date' as the index
data.set_index("Order Date", inplace=True)

# Step 5: Extract the 'Sales' column for analysis
sales_data = data["Sales"]

# Step 6: Perform the Augmented Dickey-Fuller test
adf_result = adfuller(sales_data)

# Step 7: Display the results
print("ADF Statistic:", adf_result[0])
print("p-value:", adf_result[1])
print("Critical Values:")
for key, value in adf_result[4].items():
    print(f"   {key}: {value}")

# Step 8: Interpret the results
if adf_result[1] <= 0.05:
    print("\nThe data is stationary (reject the null hypothesis).")
else:
    print("\nThe data is not stationary (fail to reject the null hypothesis).")

# Step 9: Plot the Sales data over time
plt.figure(figsize=(12, 6))
plt.plot(sales_data.index, sales_data.values, linestyle='-', color='b', label='Sales')
plt.title('Sales Over Time')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = 'C:/Users/manu/OneDrive/Documents/spin/walmart_processed_cleaned.xlsx'
data = pd.read_excel(file_path)

data['Order Date'] = pd.to_datetime(data['Order Date'])

daily_sales = data.groupby('Order Date')['Sales'].sum().reset_index()

Q1 = daily_sales['Sales'].quantile(0.25)
Q3 = daily_sales['Sales'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = daily_sales[(daily_sales['Sales'] < lower_bound) | (daily_sales['Sales'] > upper_bound)]

print("Outliers detected:")
print(outliers)
# Visualize with a scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(daily_sales['Order Date'], daily_sales['Sales'], color='blue', label='Normal Sales')  # Plot normal sales
plt.scatter(outliers['Order Date'], outliers['Sales'], color='red', label='Outliers')  # Highlight outliers
plt.title("Scatter Plot of Sales with Outliers Highlighted")
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
file_path = 'C:/Users/manu/OneDrive/Documents/spin/walmart_processed_cleaned.xlsx'
data = pd.read_excel(file_path)

# Step 2: Ensure 'Order Date' is in datetime format and sort the data
data['Order Date'] = pd.to_datetime(data['Order Date'])
data.sort_values('Order Date', inplace=True)
data.set_index('Order Date', inplace=True)
monthly_sales = data['Sales'].resample('M').sum()
model = SARIMAX(monthly_sales,
                order=(1, 1, 1),  # Non-seasonal order (p, d, q)
                seasonal_order=(1, 1, 1,12))  # Seasonal order (P, D, Q, seasonal_period)
sarima_fit = model.fit()
forecast = sarima_fit.get_forecast(steps=12)  # Forecast for the next 12 months
forecast_index = pd.date_range(monthly_sales.index[-1], periods=12, freq='M')  # Create date index for the forecast
forecast_values = forecast.predicted_mean 
conf_int = forecast.conf_int()  
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, label='Historical Sales')  # Plot historical data
plt.plot(forecast_index, forecast_values, label='Forecast', color='orange')  # Plot forecast
plt.fill_between(forecast_index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='pink', alpha=0.3, label='Confidence Interval')  # Confidence interval
plt.title('Sales Forecast with Seasonality')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

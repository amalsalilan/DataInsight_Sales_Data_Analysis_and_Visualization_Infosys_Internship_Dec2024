```python
from statsmodels.tsa.stattools import adfuller

def check_stationarity(data):
    adf_test = adfuller(data)
    print("ADF Statistic:", adf_test[0])
    print("p-value:", adf_test[1])
    print("Critical Values:", adf_test[4])
    if adf_test[1] <= 0.05:
        print("The data is stationary.")
    else:
        print("The data is not stationary. Consider differencing or transformations.")
        
check_stationarity(time_series_data['Sales'])
```

```python
# Import required libraries
import pandas as pd
from statsmodels.tsa.stattools import kpss

# Convert 'Order Date' to datetime format and sort the data by date
data['Order Date'] = pd.to_datetime(data['Order Date'])
data = data.sort_values(by='Order Date')

# Aggregate the "Sales" column by date to create a time series
# If there are multiple entries for the same date, we sum the sales
sales_time_series = data.groupby('Order Date')['Sales'].sum()

# Perform the KPSS test
# The null hypothesis (H0): The time series is stationary
# The alternative hypothesis (H1): The time series is not stationary
kpss_stat, p_value, lags, critical_values = kpss(sales_time_series)

# Print the results of the test
print("KPSS Test Results:")
print(f"KPSS Statistic: {kpss_stat}")
print(f"P-Value: {p_value}")
print(f"Lags Used: {lags}")
print("Critical Values:")
for key, value in critical_values.items():
    print(f"  {key}: {value}")

# Interpretation of Results
if p_value < 0.05:
    print("The time series is likely not stationary (reject H0).")
else:
    print("The time series is likely stationary (fail to reject H0).")
```

Identifying the missing values


```python
df.isnull().sum()
```

Not required column where data which is available doesn't contribute to the analysis and might incur in wrong analysis. Using drop function to remove the column.

```python
df.drop(columns=["Customer Age"], inplace=True)
```

```python
df.columns  #to get all the columns in the dataset
```

The code fills missing values in the `Product Base Margin` column with the most frequently occurring value (mode), ensuring data consistency for analysis.

```python
df['Product Base Margin'] = df['Product Base Margin'].fillna(df['Product Base Margin'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.info()  #to get the summary of the dtypes and the column counts
```

Identiying the relationships

```python
# Compute the correlation matrix for all numerical columns
correlation_matrix = df.select_dtypes(include= ['number']).corr()
correlation_matrix
```

```python
correlation_matrix = df[['Profit', 'Sales', 'Unit Price']].corr()
correlation_matrix
```

This code calculates the correlation matrix to analyze the strength and direction of relationships between `Profit`, `Sales`, and `Unit Price`.

1. Profit-Sales (0.58): Moderate positive correlation.  
2. Profit-Unit Price (-0.01): No significant correlation.  
3. Sales-Unit Price (0.60): Moderate positive correlation.

```python
sales_by_category = df.groupby('Customer Segment')['Sales'].agg([ 'mean', 'max', 'min'])
sales_by_category
```

Focus on the Consumer segment, as it generates the highest sales, while creating tailored strategies to unlock potential in the Corporate and Home Office segments for balanced growth.

```python
sales_summary = df[['Sales', 'Profit','Discount','Order Quantity','Shipping Cost','Unit Price']].agg(['mean', 'max', 'min'])
sales_summary

```

Sales are generally strong (mean: 1775.88), but profits can be negative (min: -14140.70). Discounts are low, and shipping costs vary significantly. Focus on controlling losses, optimizing pricing, and managing shipping costs for better profitability.

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df[['Discount', 'Order Quantity', 'Profit', 'Sales', 'Shipping Cost', 'Unit Price']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
df.groupby('Order Date')['Sales'].sum().plot()
plt.title('Total Sales Over Time', fontsize=16)
plt.xlabel('Order Date', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Product Category', y='Profit', ci=None, palette='muted')
plt.title('Profit by Product Category', fontsize=16)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Profit', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Region', hue='Ship Mode', palette='Paired')
plt.title('Ship Mode Usage by Region', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
